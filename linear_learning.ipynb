**Challenge 1**  
Generate (fake) data that is linearly related to `log(x)`. You are making this model up. It is of the form `B0 + B1*log(x) + epsilon`. (You are making up the parameters.) Simulate some data from this model. Then fit two models to it:  
quadratic (second degree polynomial)  
logarithmic (`log(x)`)  

```python
n_samples = 200
fx = range(1, n_samples+1)
#fx = np.random.uniform(1, n_samples+1, size=n_samples)#[:, np.newaxis]

def f(x):
    return 8*np.log(x) + 27

#fy = [ for x in fx]
fy = f(fx) + np.random.normal(scale=0.3, size=n_samples)#[:, np.newaxis]
print fx[:5], fy[:5]
```

```python
p = np.polyfit(fx, fy, 2)
p
```

```python
def plot_fits(fx, fy, p):
    yfit = np.polyval(p,fx)

    plt.figure()
    plt.plot(fx,fy, label='data', color="green")
    plt.plot(fx,yfit, label='fit', color="blue")
    plt.plot(fx,yfit-fy, label='var', color="red")
    plt.show()
```

```python
plot_fits(fx, fy, p)
```

```python
p1 = np.polyfit(np.log(fx), fy, 2).tolist()
p1
```

```python
plot_fits(np.log(fx),fy,p1)
```

**Challenge 2**  
Generate (fake) data from a model of the form `B0 + B1*x + B2*x^2 + epsilon`. (You are making up the parameters.)  
Split the data into a training and test set.  
Fit a model to your training set. Calculate mean squared error on your training set. Then calculate it on your test set.  
(You could use sklearn.metrics.mean_squared_error.)

```python
fx2 = sorted(np.random.uniform(1, n_samples+1, size=n_samples)[:, np.newaxis])
fy2 = [14 + 6*x + 22*(x**2) for x in fx2]
print fx2[:5], fy2[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(fx2, fy2, test_size=0.25, random_state=2)
print X_train[:5], y_train[:5]
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
```

```python
train_err = mean_squared_error(y_train, regr.predict(X_train))
test_err = mean_squared_error(y_test, regr.predict(X_test))
print "r2:", r2_score(y_test, regr.predict(X_test))
print "mse train:", train_err, "mse test:", test_err
```

**Challenge 4**  
For the data from two (above), fit a model to only the first 5 of your data points (m=5). Then to first 10 (m=10). Then to first 15 (m=15). In this manner, keep fitting until you fit your entire training set. For each step, calculate the training error and the test error. Plot both (in the same plot) over m. This is called a learning curve.

```python
tr_ers2 = []
ts_ers2 = []
ms = range(5, len(fx2), 5)
for m in ms:
    X_train = fx2[:m]
    X_test = fx2[m:]
    y_train = fy2[:m]
    y_test = fy2[m:]
    #regr = linear_model.LinearRegression()
    regr = make_pipeline(PolynomialFeatures(2), linear_model.LinearRegression())
    regr.fit(X_train, y_train)
    tr_er = mean_squared_error(y_train, regr.predict(X_train))
    ts_er = mean_squared_error(y_test, regr.predict(X_test))
    tr_ers2.append(tr_er)
    ts_ers2.append(ts_er)
ax = plt.subplot(111)
ax.yaxis.set_major_formatter(tkr.FuncFormatter(lambda x, 
                                                   pos: ('%.0f')%(x*1e-6)))
sns.plt.plot(ms, tr_ers2, color="green")
sns.plt.plot(ms, ts_ers2, color="red")

sns.plt.show()
```
